```python
import pandas as pd
import numpy as np
import jieba
import jieba.posseg as pseg
```

```python
raw = pd.read_csv('toutiao.csv')
tittle = raw.iloc[:,0]
print(raw.iloc[:,0])
t_tittle = open('tittle.txt', mode='w')
for i in tittle:
    t_tittle.write(i)
    t_tittle.write('\n')
t_tittle.close()
```

```python
text = raw.iloc[:,3]
#print(text[0])
t = text[0]
t = t.replace("\n", "")
print(t)
t_text = open('text.txt', mode = 'w')
for i in text:
    i = i.replace("\n", "")
    cut = pseg.cut(i)
    cutt = []
    for w in cut:
        if w.flag != 'x':
            cutt.append(w.word)
    t_text.write(" ".join(cutt))
    t_text.write('\n')

t_text.close()
```
